```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
# import warnings
# warnings.filterwarnings('ignore')
```

```python
titanic_data = pd.read_csv('/content/titanic.csv')
titanic_data.head()
```

```python
titanic_data.info()
```

```python
titanic_data.describe()
```

```python
titanic_data.isna()
```

```python
titanic_data.fillna(0)
```

```python
titanic_data = titanic_data.dropna(subset=['Survived'])
```

```python
Y = titanic_data['Survived']
titanic_data = titanic_data.drop('Survived', axis = 1)
```

```python
Y.head()
```

```python
titanic_data.head()
```

```python
print(titanic_data.columns)
```

```python
titanic_data = titanic_data.drop('PassengerId', axis = 1)
```

```python
titanic_data.head()
```

```python
titanic_data = titanic_data.drop('Name', axis =1)
titanic_data = titanic_data.fillna(0)
```

```python
titanic_data.head()
```

```python
gender_mapping = {'male': 0, 'female': 1}

# Create a new numeric column using .map()
titanic_data['Sex'] = titanic_data['Sex'].map(gender_mapping)
```

```python
X = titanic_data[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare']]
```

```python

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python

rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
rf_classifier.fit(X_train, y_train)
```

Final Prediction

```python
y_pred = rf_classifier.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
classification_rep = classification_report(y_test, y_pred)

print(f"Accuracy: {accuracy:.2f}")
print("\nClassification Report:\n", classification_rep)

sample = X_test.iloc[0:1]
prediction = rf_classifier.predict(sample)

sample_dict = sample.iloc[0].to_dict()
print(f"\nSample Passenger: {sample_dict}")
print(f"Predicted Survival: {'Survived' if prediction[0] == 1 else 'Did Not Survive'}")
```
